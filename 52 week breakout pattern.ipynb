```python
'''
Importing all the requierd libraries 

'''
import numpy as np
import time
import os
import pandas as pd
import yfinance as yf
from datetime import datetime, timedelta, date
```

```python
''' 
Nifty 50 stocks import 

'''
def get_50nsedata():
    data = pd.read_csv('ind_nifty50.csv')
    nse_50data = data['Symbol'].to_list()
    nse_50data = nse_50data[1:]
    return nse_50data
```

```python
'''
Nifty 200 stocks import
'''
def get_200nsedata():
    data = pd.read_csv('ind_nifty200.csv')
    nse_200data = data['Symbol'].to_list()
    return nse_200data
```

```python
'''
Download data from Yahoo finance based on stock name and date, interval
'''
def get_histdata(ticker, startdate, enddate, interval, Nse_data, prcnt_ge):
    stck_dtls_lw = []
    stck_dtls_hw = []
    tickk = ["{}.ns".format(tick) for tick in ticker]
    for line in tickk:
        symbol = line.split(",")[0]
        data = yf.download(symbol, start=startdate, end=enddate, interval=interval)
        data = data[(data[['Volume']] != 0).all(axis=1)]
        brk_out_lw = yearly_brkout_low(data, prcnt_ge, symbol, stck_dtls_lw)
        brk_out_hg = yearly_brkout_high(data, prcnt_ge, symbol, stck_dtls_hw)
    return brk_out_lw,brk_out_hg
```

```python
def yearly_brkout_low(data, prcnt_ge, symbol, stck_dtls_lw):
    year_low = round(data['Low'].min(),2)
    percent = ((prcnt_ge/100) * year_low)
    upper_limit = round(year_low + percent,2)
    lower_limit = round(year_low - percent,2)
    cur_mrk_pr = float(round(data['Close'].iloc[-1],2))
    if cur_mrk_pr > lower_limit and cur_mrk_pr < upper_limit:
        stck_dtls_lw.append(symbol.replace('.ns',''))
        #print('52 weeks low stocks are : ', symbol)
    return stck_dtls_lw
```

```python
def yearly_brkout_high(data, prcnt_ge, symbol, stck_dtls_hw):
    year_low = round(data['High'].max(),2)
    percent = ((prcnt_ge/100) * year_low)
    lower_limit = round(year_low - percent,2)
    cur_mrk_pr = float(round(data['Close'].iloc[-1],2))
    if cur_mrk_pr > lower_limit:
        stck_dtls_hw.append(symbol.replace('.ns',''))
        #print('52 weeks high stocks are : ', symbol)
    return stck_dtls_hw
```

```python
if __name__ == '__main__':
    #get the input from user nifty50 or nifty200
    Nse_data = str(input('Enter Nifty50 or Nifty200 : '))
    if Nse_data == 'Nifty50':
        Nifty_data = get_50nsedata()
    elif Nse_data == 'Nifty200':
        Nifty_data = get_200nsedata()
    else:
        pass
    prcnt_ge = int(input('enter the percentage : '))
    startdate = date.today() - timedelta(365)
    enddate = datetime.today()
    interval = '1d'
    yearly_lw, yearly_hw = get_histdata(Nifty_data, startdate, enddate, interval, Nse_data, prcnt_ge) 
    print('52 weeks low stocks are : ', yearly_lw)
    print('52 weeks high stocks are : ', yearly_hw)
    np.savetxt("52week_low.csv", yearly_lw, delimiter =", ", fmt ='% s')
    np.savetxt("52week_high.csv", yearly_hw, delimiter =", ", fmt ='% s')
```
